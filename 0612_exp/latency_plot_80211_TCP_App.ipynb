```python
import csv

def read_int_from_file(file_location):
    numbers = []
    with open(file_location) as tsv:
        for column in zip(*[line for line in csv.reader(tsv, dialect="excel-tab")]):
            for num in column:
                numbers.append(num)
    while '' in numbers:
        numbers.remove('')

    print(file_location + " has " + str(len(numbers)) + " of entries.")
    numbers_int = map(int, numbers)
    numbers_int_list = list(numbers_int)
    
    return numbers_int_list

def read_float_from_file(file_location):
    numbers = []
    with open(file_location) as tsv:
        for column in zip(*[line for line in csv.reader(tsv, dialect="excel-tab")]):
            for num in column:
                numbers.append(num)
    while '' in numbers:
        numbers.remove('')
    
    print(file_location + " has " + str(len(numbers)) + " of entries.")
    numbers_float = map(float, numbers)
    numbers_float_list = list(numbers_float)
    
    return numbers_float_list

interference_no_list = read_int_from_file("time_record/no_interf_time_record_raw.txt")
#interference_no_VI_list = read_int_from_file("time_record/no_interf_VI_time_record_raw.txt")
#interference_no_VO_list = read_int_from_file("time_record/no_interf_VO_time_record_raw.txt")
interference_max_list = read_int_from_file("time_record/max_interf_time_record_raw.txt")
interference_max_VI_list = read_int_from_file("time_record/max_interf_VI_time_record_raw.txt")
interference_max_VO_list = read_int_from_file("time_record/max_interf_VO_time_record_raw.txt")

interference_no_list_chunks = [interference_no_list[x:x+20] for x in range(0, len(interference_no_list), 20)]
#interference_no_VI_list_chunks = [interference_no_VI_list[x:x+20] for x in range(0, len(interference_no_VI_list), 20)]
#interference_no_VO_list_chunks = [interference_no_VO_list[x:x+20] for x in range(0, len(interference_no_VO_list), 20)]
interference_max_list_chunks = [interference_max_list[x:x+20] for x in range(0, len(interference_max_list), 20)]
interference_max_VI_list_chunks = [interference_max_VI_list[x:x+20] for x in range(0, len(interference_max_VI_list), 20)]
interference_max_VO_list_chunks = [interference_max_VO_list[x:x+20] for x in range(0, len(interference_max_VO_list), 20)]

interference_no_list_TCP = read_float_from_file("time_record/no_evcc_TCP_UDP.txt")
#interference_no_VI_list_TCP = read_float_from_file("time_record/no_VI_evcc_TCP_UDP.txt")
#interference_no_VO_list_TCP = read_float_from_file("time_record/no_VO_evcc_TCP_UDP.txt")
interference_max_list_TCP = read_float_from_file("time_record/max_evcc_TCP_UDP.txt")
interference_max_VI_list_TCP = read_float_from_file("time_record/max_VI_evcc_TCP_UDP.txt")
interference_max_VO_list_TCP = read_float_from_file("time_record/max_evcc_VO_TCP_UDP.txt")

interference_no_list_TCP_chunks = [interference_no_list_TCP[x:x+20] for x in range(0, len(interference_no_list_TCP), 20)]
#interference_no_VI_list_TCP_chunks = [interference_no_VI_list_TCP[x:x+20] for x in range(0, len(interference_no_VI_list_TCP), 20)]
#interference_no_VO_list_TCP_chunks = [interference_no_VO_list_TCP[x:x+20] for x in range(0, len(interference_no_VO_list_TCP), 20)]
interference_max_list_TCP_chunks = [interference_max_list_TCP[x:x+20] for x in range(0, len(interference_max_list_TCP), 20)]
interference_max_VI_list_TCP_chunks = [interference_max_VI_list_TCP[x:x+20] for x in range(0, len(interference_max_VI_list_TCP), 20)]
interference_max_VO_list_TCP_chunks = [interference_max_VO_list_TCP[x:x+20] for x in range(0, len(interference_max_VO_list_TCP), 20)]

interference_no_list_80211 = read_float_from_file("time_record/no_interf_80211.txt")
interference_no_list_80211_chunks = [interference_no_list_80211[x:x+17] for x in range(0, len(interference_no_list_80211), 17)]

interference_max_list_80211 = read_float_from_file("time_record/max_interf_80211.txt")
interference_max_list_80211_chunks = [interference_max_list_80211[x:x+18] for x in range(0, len(interference_max_list_80211), 18)]

interference_max_VO_list_80211 = read_float_from_file("time_record/max_VO_80211.txt")
interference_max_VO_list_80211_chunks = [interference_max_VO_list_80211[x:x+18] for x in range(0, len(interference_max_VO_list_80211), 18)]

interference_max_VI_list_80211 = read_float_from_file("time_record/max_VI_80211.txt")
interference_max_VI_list_80211_chunks = [interference_max_VI_list_80211[x:x+20] for x in range(0, len(interference_max_VI_list_80211), 20)]
```

```python
#Calculate the dropping percent of the application layer round trip latency
from statistics import mean

stage_to_string_14 = [
    "Start <-> SECC Discovery",
    "SECC Discovery <-> Session Initialized",
    "Session Initialized <-> SupportedAppProtocol",
    "SupportedAppProtocol <-> SessionSetup",
    "SessionSetup <-> ServiceDiscovery",
    "ServiceDiscovery <-> PaymentServiceSelection",
    "PaymentServiceSelection <-> PaymentDetail",
    "PaymentDetail <-> Authorization",
    "Authorization <-> ChargeParameterDiscovery",
    "ChargeParameterDiscovery <-> Start Charging",
    "Start <-> Session Initialized",
    "Start <-> Start Charging",
    "During Charging(related to config.)",
    "Stop Charging <-> Session Close"
    ]

def latencyDropPercent(list_before, list_after):
    if(len(list_before) != len(list_after)):
        print("Length doesn't conform!")
        return
    
    for i in range(0, len(list_before)):
        dropPercent = (mean(list_before[i])-mean(list_after[i]))/mean(list_before[i])
        print(stage_to_string_14[i] + ": " + str(dropPercent * 100) + "%")

latencyDropPercent(interference_max_list_chunks, interference_max_VI_list_chunks)
```
